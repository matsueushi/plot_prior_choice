# Plots of Prior Choice Recommendations
https://github.com/stan-dev/stan/wiki/Prior-Choice-Recommendations

```julia
using Distributions
using Plots
```

## Priors

```julia
xs = 0:0.1:50
plot(xs, pdf.(Normal(0, 1e6), xs), label = "Super-vague: Normal(0, 1e6)", linewidth = 2)
plot!(xs, pdf.(Normal(0, 10), xs), label = "Weakly informative prior, very weak: Normal(0, 10)", linewidth = 2)
plot!(xs, pdf.(Normal(0, 1), xs), label = "Generic weakly informative prior: Normal(0,1)", linewidth = 2)
```

```julia
xs = 0:0.01:2
plot(xs, pdf.(Normal(0, 1e6), xs), label = "Super-vague: Normal(0, 1e6)", linewidth = 2)
plot!(xs, pdf.(Normal(0, 10), xs), label = "Weakly informative prior, very weak: Normal(0, 10)", linewidth = 2)
plot!(xs, pdf.(Normal(0, 1), xs), label = "Generic weakly informative prior: Normal(0,1)", linewidth = 2)
plot!(xs, pdf.(Normal(0.4, 0.2), xs), label = "Specific informative prior: Normal(0.4, 0.2)", linewidth = 2)
```

## Genaral principals

```julia
xs = -1:0.01:2
plot(xs, pdf.(Uniform(0, 1), xs), label = "Uniform(0, 1)", linewidth = 2)
plot!(xs, pdf.(Normal(0.5, 0.5), xs), label = "Normal(0.5, 0.5)", linewidth = 2)
```

```julia
xs = 0:1000
plot(xs, pdf.(Uniform(0, 100), xs), label = "Uniform(0, 100)", linewidth = 2)
plot!(xs, pdf.(Uniform(0, 1000), xs), label = "Uniform(0, 1000)", linewidth = 2)
```

```julia
xs = 0:0.1:50
plot(xs, pdf.(Uniform(0, 100), xs), label = "Uniform(0, 100)", linewidth = 2)
plot!(xs, pdf.(Uniform(0, 1000), xs), label = "Uniform(0, 1000)", linewidth = 2)
# Exponential(theta) in Distribution.jl = exponential(1/theta) in Stan
plot!(xs, pdf.(Exponential(10), xs), label = "Exponential(10)", linewidth = 2)
plot!(xs, pdf.(Truncated(Normal(0, 10), 0, Inf), xs), label = "HalfNormal(0, 10)", linewidth = 2)
plot!(xs, pdf.(Truncated(Cauchy(0, 5), 0, Inf), xs), label = "HalfCauchy(0, 5)", linewidth = 2)
```

```julia
xs = 0:0.1:10
plot(xs, pdf.(Uniform(0, 100), xs), label = "Uniform(0, 100)", linewidth = 2)
plot!(xs, pdf.(Uniform(0, 1000), xs), label = "Uniform(0, 1000)", linewidth = 2)
plot!(xs, pdf.(Exponential(10), xs), label = "Exponential(10)", linewidth = 2)
plot!(xs, pdf.(Truncated(Normal(0, 10), 0, Inf), xs), label = "HalfNormal(0, 10)", linewidth = 2)
plot!(xs, pdf.(Truncated(Cauchy(0, 5), 0, Inf), xs), label = "HalfCauchy(0, 5)", linewidth = 2)
plot!(xs, pdf.(Truncated(Normal(0, 1), 0, Inf), xs), label = "HalfNormal(0, 1)", linewidth = 2)
plot!(xs, pdf.(Truncated(TDist(3), 0, Inf), xs), label = "HalfTDist(3)", linewidth = 2)
```
